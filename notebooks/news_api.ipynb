# Collecting news articles from the News API

```python
import pandas as pd
import requests
```

```python
def get_headlines(url):
    page = requests.get(url)
    headlines = page.json()
    return headlines
```

```python
url = 'https://newsapi.org/v2/everything?q=cryptocurrency&sortBy=publishedAt&apiKey=e3e38edf30814dc18ef7f21a7fb9e775'
```

```python
headlines = get_headlines(url)
```

```python
headlines
```

```python
articles = headlines['articles']
articles
```

```python
len(articles)
```

```python
# Construct dataframe with article information
dfs = []
for article in articles:
    time = article['publishedAt']
    title = article['title']
    source = article['source']['name']
    dfs.append([time, title, source])
data = pd.DataFrame(dfs, columns=['Date', 'Headline', 'Source'])
```
